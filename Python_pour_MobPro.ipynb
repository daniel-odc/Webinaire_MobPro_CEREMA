![Chargement depuis un fichier local](./img/cell1_chargement_local.png)

```python
# Préparation de la donnée
import pandas as pd  # Lancer avant `pip install pandas`

# # Chargement des bases:

# # # Option 1: Chargement depuis un fichier local téléchargé depuis le site de l'INSEE:
# ~1min30 de chargement en mémoire (hors temps de téléchargement manuel)

# # (adaptez votre répertoire local où vous avez sauvegardé les fichiers)
df_mobpro = pd.read_csv(r'c:\Users\dmarin\Downloads\FD_MOBPRO_2018.zip', sep=';', low_memory=False)
df_mobsco = pd.read_csv(r'c:\Users\dmarin\Downloads\FD_MOBSCO_2018.zip', sep=';', low_memory=False)

print(f'df_mobpro: {len(df_mobpro.index):,} enregistrements\n')  # 7,932,895 enregistrements
display(df_mobpro.head())

print(f'df_mobsco: {len(df_mobsco.index):,} enregistrements\n')  # 4,747,781 enregistrements
display(df_mobsco.head())

```

![Chargement via l'API Pynsee](./img/cell2_chargement_api.png)

```python
# # # Option 2: Chargement depuis l'API Pynsee: ~3min: téléchargement + chargement en mémoire
import pynsee # Lancer avant `pip install pynsee`

df_mobpro = pynsee.download.download_file('RP_MOBPRO_2018')
df_mobsco = pynsee.download.download_file('RP_MOBSCO_2018')

print(f'df_mobpro: {len(df_mobpro.index):,} enregistrements\n')  # 7,932,895 enregistrements
print(f'df_mobsco: {len(df_mobsco.index):,} enregistrements\n')  # 4,747,781 enregistrements
```

![Sélection des colonnes et dictionnaires de valeurs](./img/cell3_selection_colonnes.jpg)

```python
# # Extraction des champs pertinents et réagencement des bases:

# Sélection des colonnes qui nous intéressent:
col_pro = ['COMMUNE', 'ARM', 'DCLT', 'IPONDI', 'TP', 'TRANS']   # mobpro
col_sco = ['COMMUNE', 'ARM', 'DCETUF', 'IPONDI']                # mobsco

# Renommage des colonnes de façon plus explicite:
dict_rename = {
    'COMMUNE':  'depuis',           # commune d'origine dans les fichiers mobpro et mobsco (hors Paris, Lyon, Marseille)
    'ARM':      'arrondissement',   # arrondissement d'origine pour Paris, Lyon, Marseille dans les fichiers mobpro et mobsco
    'DCLT':     'vers',             # commune de destination dans les fichiers mobpro
    'DCETUF':   'vers',             # commune de destination dans les fichiers mobsco
    'IPONDI':   'poids',            # pondération de l'individu dans les fichiers mobpro et mobsco
    'TP':       'temps_partiel',    # indication de temps partiel / temps plein dans les fichiers mobpro
    'TRANS':    'mode'              # mode de transport pour aller travailler dans les fichiers mobpro
}

# Dictionnaires de valeurs de champs:

# Modes de transport:
dict_mode = {
    1: 'pas_de_transport',
    2: 'marche',          # Marche à pied(ou rollers, patinette)
    3: 'vélo',            # Vélo(y compris à assistance électrique) (apparu depuis les bases 2018)
    4: '2rm',             # Deux - roues motorisé
    5: 'vp',              # Voiture, camion, fourgonnette
    6: 'tc',              # Transports en commun
    'Z': 'nan'}             # non resnsigné

#   Temps partiel vs temps plein:
dict_tp = {
    1: 1,     # 'Temps complet'
    2: 0.5}   # 'Temps partiel: on suppose un mi-temps
```

```python
# # Sélection et renommage des colonnes pertinentes:

# Sélection des colonnes:
df_pro = df_mobpro[col_pro].rename(columns=dict_rename)
df_sco = df_mobsco[col_sco].rename(columns=dict_rename)

# Remplacement des valeurs par des valeurs plus explicites:
df_pro['mode'] = df_pro['mode'].replace(dict_mode)
df_pro['temps_partiel'] = df_pro['temps_partiel'].replace(dict_tp)

print(f'df_pro:')
display(df_pro.head())

print(f'df_sco:')
display(df_sco.head())
```

![Gestion des arrondissements et pondération temps partiel](./img/cell5_arrondissements_poids.jpg)

```python
# # Remplacement des communes par les arrondissements pour Paris Lyon, Marseille
df_pro.loc[(df_pro['arrondissement'] != 'ZZZZZ'), 'depuis'] = df_pro.loc[(df_pro['arrondissement'] != 'ZZZZZ'), 'arrondissement']
df_sco.loc[(df_sco['arrondissement'] != 'ZZZZZ'), 'depuis'] = df_sco.loc[(df_sco['arrondissement'] != 'ZZZZZ'), 'arrondissement']


# Pondération du poids par le temps partiel
df_pro['poids'] = df_pro['poids'] * df_pro['temps_partiel']

# Retrait des colonnes arrondissement et temps partiel:
df_pro.drop(columns=['arrondissement', 'temps_partiel'], inplace=True)
df_sco.drop(columns=['arrondissement'], inplace=True)

print(f'df_pro:')
display(df_pro.head())

print(f'df_sco:')
display(df_sco.head())
```

![Agrégation groupby et pivot par mode de transport](./img/cell6_groupby_pivot.jpg)

```python
# # Regroupement des enregistrements par 'depuis'-'vers':

# Prise en compte des temps partiels
df_pro = df_pro.groupby(by=['depuis', 'vers', 'mode'], as_index=False)['poids'].sum()
df_sco = df_pro.groupby(by=['depuis', 'vers'], as_index=False)['poids'].sum()

# Pivot par mode de transport: une colonne par mode:
df_pro_pivot = pd.pivot_table(df_pro, index=['depuis', 'vers'], columns='mode', fill_value=0)

# Transforme les noms de colonnes de 'poids marche' à simplement 'marche':
df_pro_pivot.columns = [f"{col[1]}" for col in df_pro_pivot.columns]  # removes MultiIndex

print(f'df_pro:')
display(df_pro_pivot.head())
```

```python
# ajout total flux pro:
df_pro_pivot['tot_pro'] = 0
for mode in dict_mode.values():
    if mode in df_pro_pivot.columns:
        df_pro_pivot['tot_pro'] = df_pro_pivot['tot_pro'] + df_pro_pivot[mode]

# renomme total flux scolaires:
df_sco.rename(columns={'poids': 'tot_sco'}, inplace=True)

print(f'df_pro:')
display(df_pro_pivot.head())

print(f'df_sco:')
display(df_sco.head())
```

![Fusion des tables professionnelles et scolaires](./img/cell8_fusion.png)

```python
# # Fusion des tables scolaires et professionnelles:

# Fusion des tables sur les depuis-vers:
df_mob = df_pro_pivot.merge(df_sco, on=['depuis', 'vers'], how='left')

print(f'df_mob: {len(df_mob.index):,} enregistrements')
display(df_mob.head())
```
